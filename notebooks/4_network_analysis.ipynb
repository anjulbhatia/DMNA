```python
import networkx as nx
import pandas as pd

path_to_dir = '../data/processed_data/'
```

```python
# Load the necessary data (Replace with your actual data loading mechanism)
stations = pd.read_csv(path_to_dir + 'stations.csv')  # Contains station info like id, name, latitude, longitude, etc.
routes = pd.read_csv(path_to_dir + 'routes.csv')  # Contains route info like route_id, start station, end station
stop_times = pd.read_csv(path_to_dir + 'stop_times.csv')  # Contains stop sequence and distances between stations
shapes = pd.read_csv(path_to_dir + 'shapes.csv') # Contains info like shape_id, shape_pt_sequence etc.
```

```python
stations.head()
```

```python
# Create a graph object
metro_graph = nx.Graph()

# Adding nodes (stations)
for index, row in stations.iterrows():
    station_id = row['station_id']
    station_name = row['station_name']
    metro_line = row['metro_line']
    latitude = row['latitude']
    longitude = row['longitude']
    station_type = row['station_type']  # Elevated, Underground, At-Grade

    # Adding station as a node with attributes
    metro_graph.add_node(station_id, name=station_name, line=metro_line, 
                         lat=latitude, lon=longitude, type=station_type)

```

```python
complete_routes = routes[routes['route_type'] == 'Complete']['route_id'].tolist()

# Add Aqua line to the complete_routes
complete_routes.extend([34, 35])

print(complete_routes)
```

```python
complete_shapes = shapes[shapes['route_id'].isin(complete_routes)].sort_values(by=('route_id'))

complete_shapes.head()
```
