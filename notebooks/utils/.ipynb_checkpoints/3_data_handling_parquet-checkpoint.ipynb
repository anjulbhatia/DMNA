## [Aux] Parquet File Data Handling

#### 1 Import necessary modules and libraries and initiate the project

```python
import os
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from IPython.display import display, Markdown
import warnings
warnings.filterwarnings('ignore')
```

#### 2 Load the Parquet Data

```python
metro_data = pd.read_parquet("../data/processed_data/metro_network_data.parquet")
stops = pd.read_csv("../data/cleaned_data/stops.csv")
```

```python
stop_names = stops[['stop_name']].set_index(stops.stop_id)
stop_name_dict = stops.set_index('stop_id')['stop_name'].to_dict()

display(stop_name_dict)
```

```python
metro_data_with_stop_names = metro_data.copy()
metro_data_with_stop_names['stop_name'] = metro_data_with_stop_names['stop_id'].map(stop_name_dict)
```

```python
metro_data_with_stop_names.drop_duplicates()
```
